# Check Distributions

[Check this](https://numpy.org/doc/stable/reference/random/generated/numpy.random.pareto.html)

```python
import numpy as np
import pandas as pd

from numpy.random import MT19937
from numpy.random import RandomState, SeedSequence

nbr_records = 100
rs = RandomState(MT19937(SeedSequence(42)))
```

## Binomial

```python
dfData = pd.DataFrame({'app_id': np.arange(0, nbr_records)})
data = (np.random.binomial(1, .5, nbr_records) + 1) * 3
data = pd.Series(data)
dfData = dfData.assign(dist1=data)

data = (np.random.binomial(10, .5, nbr_records) + 1) * 2
data = pd.Series(data)
dfData = dfData.assign(dist2=data)

data = (np.random.binomial(10, .2, nbr_records) + 1) * 1
data = pd.Series(data)
dfData = dfData.assign(dist3=data)

dfData = dfData.drop('app_id', axis=1)
dfData.plot.kde() #hist, kde, line
print( dfData.describe() )
del dfData, data
```
